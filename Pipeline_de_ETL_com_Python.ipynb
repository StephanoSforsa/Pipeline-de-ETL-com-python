```python
#Bibliotecas utilizadas:
import pandas as pd
import random
```

## **E**xtract

Criando uma base de dados fictícia no formato CSV e extraindo os dados  de usuário a partir do arquivo.

```python
#Criando uma base de dados:

#Fixando a seed, para garantir que o mesmo resultado será gerado ao rodar o notebook novamente:
random.seed(3)

# Listas de nomes e sobrenomes brasileiros (mais comuns):
nomes = ['João', 'Maria', 'Pedro', 'Ana', 'Lucas', 'Isabella', 'Luiza', 'Gustavo', 'Larissa', 'Rafael']
sobrenomes = ['Silva', 'Santos', 'Oliveira', 'Pereira', 'Ferreira', 'Rodrigues', 'Almeida', 'Lima', 'Souza', 'Costa']

# Número de nomes que serão gerados:
num_nomes = 10

# Gerando nomes aleatórios:
nomes_aleatorios = [random.choice(nomes) + ' ' + random.choice(sobrenomes) for _ in range(num_nomes)]

# Gerando os IDs:
user_ids = range(10)

# Gerando os pesos:
pesos = [random.uniform(70, 110) for _ in range(10)]

# Gerando as alturas:
alturas = [random.uniform(150, 190) for _ in range(10)]

pd.DataFrame({'UserID':user_ids,'User_name':nomes_aleatorios,'Peso':pesos, 'Altura':alturas}).to_csv('dados_clientes.csv')
```

Extraindo os dados:

```python
df = pd.read_csv('dados_clientes.csv', usecols=['UserID', 'User_name', 'Peso', 'Altura'])
df.sort_values('User_name')
```

## **T**ransform

Foram feitas as seguintes trasnformações:

1. Transformação da coluna de altura do formato de cm para metros.
2. Arredondamento dos valores das colunas peso e altura para 2 casas decimais.
3. Cálculo do IMC para cada cliente.
4. Classificando a faixa que cada cliente se encaixa.

```python
# 1 - Transformando a altura em metros:
df['Altura'] = df['Altura']/100
```

```python
# 2 - Diminuindo o numero de casas deciamis:
df['Altura'] = df['Altura'].apply(lambda x: round(x, 2))
df['Peso'] = df['Peso'].apply(lambda x: round(x, 2))
```

```python
# 3 - Calculando o IMC:
df['IMC'] = df['Peso'] / (df['Altura'] ** 2)
```

```python
# 4 - Classificando a faixa que cada cliente se encaixa:
# Função para mapear valores de IMC para categorias:
def categorizar_imc(imc):
    if imc < 18.5:
        return 'Abaixo do peso'
    elif 18.5 <= imc < 25.0:
        return 'Faixa saudável'
    elif 25.0 <= imc < 30.0:
        return 'Sobrepeso'
    elif 30.0 <= imc < 35.0:
        return 'Obesidade grau 1'
    elif 35.0 <= imc < 40.0:
        return 'Obesidade grau 2'
    else:
        return 'Obesidade grau 3'

df['Categoria IMC'] = df['IMC'].apply(categorizar_imc)
```
